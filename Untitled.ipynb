```python
class KMeanClustering:
    def __init__(self, k=5, threshold=0.0001, inertia=30000):
        self.k = k
        self.centroids = None
        self.iterations_run = 0  # Track how many iterations actually ran
        self.inertia_ = []       # Track loss function values
        self.threshold = threshold
        self.inertia = inertia
    
    @staticmethod
    def euclidean_distance(data_point, centroid):
        return np.sqrt(np.sum((centroid-data_point)**2, axis=1))

    def initialize_centroids(self, X):
        n_samples, n_features = X.shape
        centroids = np.zeros((self.k, n_features))
    
        #Randomly pick the first centroid
        centroids[0] = X[np.random.randint(0, n_samples)]
    
        #Pick remaining centroids
        for i in range(1, self.k):
            distances = np.min(np.square(np.linalg.norm(X[:, np.newaxis] - centroids[:i], axis=2)), axis=1)
            probabilities = distances / distances.sum()
            cumulative_probs = np.cumsum(probabilities)
            r = np.random.rand()
            index = np.searchsorted(cumulative_probs, r)
            centroids[i] = X[index]
    
        return centroids

    def initialize_centroids_quantile_based(self, X):
        
        n_samples, n_features = X.shape
        centroids = np.zeros((self.k, n_features))
        
        # For each feature, divide into k quantiles and sample from each
        for i in range(self.k):
            quantile = (i + 1) / (self.k + 1)  # Avoid 0 and 1 quantiles
            for j in range(n_features):
                centroids[i, j] = np.quantile(X[:, j], quantile)
        
        return centroids
        
    def fit(self, X, max_iterations = 100, initialization_strategy="Random", convergence_criteria = 'Centroid Shift'):

        self.centroids = 0
        if(initialization_strategy== 'Random'):
            self.centroids = np.random.uniform(np.amin(X, axis=0), np.amax(X, axis=0), size=(self.k, X.shape[1]))  # rando,m 
        if(initialization_strategy== 'KMeans++'):
            self.centroids = self.initialize_centroids(X)
        if(initialization_strategy== 'Q'):
            self.centroids = self.initialize_centroids_quantile_based(X)
        
        for _ in range(max_iterations):
            self.iterations_run = self.iterations_run + 1
            y= []
            for data_point in X:
                distances = KMeanClustering.euclidean_distance(data_point, self.centroids)
                cluster_num = np.argmin(distances)
                y.append(cluster_num)

            y=np.array(y)

            cluster_indices = []
            for i in range(self.k):
                cluster_indices.append(np.argwhere(y == i))

            cluster_centers = []

            for i, indices in enumerate(cluster_indices):
                if len(indices) == 0:
                    cluster_centers.append(self.centroids[i])
                else:
                    cluster_centers.append(np.mean(X[indices.flatten()], axis=0))

            loss = np.sum([
                    np.min(KMeanClustering.euclidean_distance(x, np.array(cluster_centers))) ** 2 for x in X
            ])
            self.inertia_.append(loss)

            # Convergence criteria
            if convergence_criteria == 'Max Iterations':
                if i == max_iterations - 1:  # Final iteration
                    break
            elif convergence_criteria == 'Centroid Shift':
                if np.max(np.linalg.norm(self.centroids - np.array(cluster_centers),axis=1)) < self.threshold:
                    break
            elif convergence_criteria == 'Loss Function Threshold':
                if len(self.inertia_) > 1 and abs(self.inertia_[-2] - self.inertia_[-1]) < self.threshold:
                    break
            
            self.centroids = np.array(cluster_centers)


        return y

```

```python
# Instantiate and fit the model
kmeans = KMeanClustering(k=3)
labels = kmeans.fit(scaled_data, 100, initialization_strategy="KMeans++")

data = scaled_data

plt.scatter(data[:,0], data[:,1], c=labels)
plt.scatter(kmeans.centroids[:,0], kmeans.centroids[:,1], c=range(len(kmeans.centroids)), marker='*', s=200)
plt.show()
inertia = 0
for i in range(len(kmeans.inertia_)):
    inertia = inertia + float(kmeans.inertia_[i])
print("Iterations run:", kmeans.iterations_run)
print("Final total loss fuction:", inertia)
```

# Determining the Optimal K

```python
def detect_elbow(K_values, losses):
    # Convert to numpy arrays
    K_values = np.array(K_values)
    losses = np.array(losses)
    
    # Line connecting first and last points
    point1 = np.array([K_values[0], losses[0]])
    point2 = np.array([K_values[-1], losses[-1]])
    
    # Compute distances from each point to the line
    distances = []
    for i in range(len(K_values)):
        p = np.array([K_values[i], losses[i]])
        # Distance from point to line formula
        dist = np.abs(np.cross(point2-point1, point1-p)) / np.linalg.norm(point2-point1)
        distances.append(dist)
    
    distances = np.array(distances)
    
    # The elbow is the K with maximum distance
    optimal_k = K_values[np.argmax(distances)]
    return optimal_k

```

```python


losses = []
K_values = range(2, 16)

for k in K_values:
    kmeans = KMeanClustering(k=k)
    labels = kmeans.fit(scaled_data, max_iterations=10, initialization_strategy='KMeans++', convergence_criteria = 'Centroid Shift')
    losses.append(kmeans.inertia_[-1])


plt.plot(K_values, losses, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Loss (Inertia)')
plt.title('Elbow Method for Optimal K')
plt.show()

optimal_k = detect_elbow(K_values, losses)
print(f"Automatically detected optimal K: {optimal_k}")
```
